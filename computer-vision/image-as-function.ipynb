### Images as Functions I(x, y)

- a 2dimensional array of numbers, ranging from some minimum to some maximum
- a function of I(x, y)
- something generated by a camera

> f : [a, b] x [c, d] --> [min, max]


### Color Images as Functions

#### A color image is 3 functions "stacked" together, "vector-valued function"

> f(x, y) = [ r(x, y),
              g(x, y),
              b(x, y)
            ]
            
#### 3 channel color image is a mapping from a 2-dimensional space of locations to a 3 dimensional space of color intensity values

### Digital Images "discrete images"

- sample

> 2d space on a regular grid

- quantize

> each sample (round to the nearest integer)

#### "discrete picture elements at specific locations"
#### "16bit images" / "12bit images"
