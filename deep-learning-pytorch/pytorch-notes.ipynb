### Packages

 - __torch.nn__
 
 > a subpackage that containsmodules and extensible classes for building NNs
 
 - __torch.autograd__
 
 > supports all the differentiable Tensor operations in PyTorch
 >
 > automatically calculating gradients of tensors
 
 - __torch.nn.functional__ (loss/activation/convolution)
 
 > a functional interface that contains typical operations used for buildings NNs
 
 - __torch.optim__
 
 > contains standard optimization operations like SGD and Adam
 
 - __torch.utils__
 
 > contains utility classes like datasets and data loaders. make data preprocessing easier
 
 - __torchvision__
 
 > provides access to popular datasets, model architectures, and image transformations for computer vision
 
 - __nn.dropout__

### Tensor is the core data structure in PyTorch
 - weights.reshape(a, b)
 
 > will return a new tensor with the same data as weights w/size (a, b) sometimes, and sometimes a clone, as in: it copies data to another part of memory
 
 - __weights.resize_(a, b)__
 
 > returns the same tensor with a different shape, underscore indicates "inplace"
 
 - __weights.view(a, b)__
 
 > will return a new tensor w/ the same data as weights w/size(a, b)
 
 - __torch.mm__
 
 > matrix multiplication, very specific, tensors need to be same shape
 
 - __torch.matmul__
 
 > supports broadcasting, tensors need to be same shape
